# 1. Obtener los datos de entrada. Guardar todas las im√°genes de entrada dentro de un array

```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage.transform import iradon # Mousequerramienta misteriosa

tamano_detector = 256 # en pixeles
numero_imagenes = 349 # Numero de imagenes "verticales"
angulo = 0.6 # angulo entre imagen e imagen ()

IMGs = [] # lista que va a contener todas las imagenes que vamos a cargar. Cada componente es una imagen. 

for i in range(numero_imagenes):
    IMG = np.genfromtxt("Imagenes/CORR/corr_{}.txt".format(i))
    IMGs.append(IMG)

IMGs = np.array(IMGs)
```

```python
plt.imshow(IMGs[300], cmap="gray", vmin=5 , vmax=7)
plt.colorbar()
plt.show()
```

# 2. Obtener sinogramas

## Previo: Obtener una sola rodaja primero. Luego todas... no nos precipitemos

```python
banditas = []

for imagen in IMGs:
    bandita = np.sum(imagen[30:60, :], axis=0) # Sumamos por columnas la bandita imagen[30:60, :]
    banditas.append(bandita)

vector_banditas = np.array(banditas)
sinograma = np.transpose(vector_banditas)
```

```python
plt.imshow(sinograma, cmap="gray")
plt.xlabel("Numero de imagenes verticales")
plt.ylabel("Altura del detector")
plt.colorbar()
plt.show()
```

# 3. Obtener la rodaja

```python
# Ejemplo de np.linspace
x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x,y)
```

```python
theta = np.linspace(0.0, angulo*numero_imagenes, numero_imagenes) # np.linspace(a,b,n) crea un vector con n valores (equidistanciados) entre a y b. En este caso, nos da la informaci√≥n de cuantas imagenes se tomaron y en qu√© rango de √°ngulos se tomaron

rodaja = iradon(sinograma, theta=theta, filter_name='hann') # La rodaja es la transformada inversa de rad√≥n del sinograma. Toma como entradas el sinograma, el vector con los angulos que definimos previamente y (opcionalmente) un filtro que desees aplicar
```

```python
plt.imshow(rodaja, cmap="gray", vmin=0.6 , vmax= 1.0)
plt.colorbar()
plt.show()
```

# Unamos todas las partes en una sola celda

```python
import numpy as np
import matplotlib.pyplot as plt 
from skimage.transform import iradon 

###################### Definir variables de interes ####################################################

tamano_detector = 256 # en pixeles
numero_imagenes = 349 # Numero de imagenes "verticales"
angulo = 0.6 # angulo entre imagen e imagen

####################### Cargar y guardar las imagenes verticales #######################################

IMGs = [] # lista que va a contener todas las imagenes que vamos a cargar. Cada componente es una imagen. 

for i in range(numero_imagenes):
    IMG = np.genfromtxt("Imagenes/CORR/corr_{}.txt".format(i))
    IMGs.append(IMG)

IMGs = np.array(IMGs) # Convertimos a IMGs de lista a array

####################### Formar el sinograma de una sola bandita ##############################################

banditas = []

for imagen in IMGs:
    bandita = np.sum(imagen[30:60, :], axis=0) # Sumamos por columnas la bandita imagen[30:60, :]
    banditas.append(bandita)

vector_banditas = np.array(banditas)
sinograma = np.transpose(vector_banditas)

############# Aplicamos la transformada inversa de Radon al sinograma para obtener la rodaja ####################

theta = np.linspace(0.0, angulo*numero_imagenes, numero_imagenes) 

rodaja = iradon(sinograma, theta=theta, filter_name='hann') 

############################## Graficamos la rodaja (Imagen 2D) ##################################################

plt.imshow(rodaja, cmap="gray", vmin=0.6 , vmax= 1.0)
plt.colorbar()
plt.show()
```
